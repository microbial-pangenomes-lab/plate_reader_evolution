```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('ticks',
              rc={'axes.facecolor': (0, 0, 0, 0)})
sns.set_context('talk')

from matplotlib import rcParams
rcParams['font.family'] = 'sans-serif'
rcParams['figure.dpi'] = 96
```

```python
import numpy as np
import pandas as pd
```

```python
m = pd.read_csv('my_randomization.tsv', sep='\t', header=None)
m.columns = ['r1', 'c1', 'r2', 'c2']
m = m.sort_values(['r1', 'c1'])
border = m[(m['r1'].isin(['A', 'H'])) |
           (m['c1'].isin([1, 12]))].index
middle = m.loc[m.index.difference(border)].sort_values(['r1', 'c1']).index
m.loc[border, 'v1'] = np.arange(10, len(border) + 10)
m.loc[middle, 'v1'] = -np.arange(10, len(middle) + 10)
```

```python
sns.heatmap(m.pivot_table(index='r1', columns='c1', values='v1'),
            cmap='vlag')

plt.title('original plate');
```

```python
sns.heatmap(m.pivot_table(index='r2', columns='c2', values='v1'),
            cmap='vlag')

plt.title('randomized plate');
```
